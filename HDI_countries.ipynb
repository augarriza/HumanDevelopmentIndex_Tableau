```python
df_pop["Above_below"] = np.where(df_pop["HDI"] > 0.7, "Above", "Below")
```

```python
df_pop.head(2)
```

```python
df_ab = (df_pop.groupby("Above_below")
         .agg({"Total Population (millions) 2015":"sum"
               ,"Total Population (millions) 2030":"sum"}))
```
